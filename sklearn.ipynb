<h2>Simple Regression</h2>

```python
data = {
    'Age'        : [2, 8, 10, 14, 18, 22, 28, 35, 41, 47, 54, 63, 77, 85, 99, 110],
    'Experience' : [10, 60, 110, 135, 200, 218, 295, 350, 400, 450, 568, 637, 700, 904, 976, 1210]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
X = df[['Age']]
y = df['Experience']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f'Intercept: {model.intercept_}')
print(f'Slope: {model.coef_[0]}')

plt.scatter(X, y, color='blue', label='Actual Experience')
plt.plot(X_test, y_pred, color='black', linewidth=1, label='Predicted Experience')
plt.xlabel('Age')
plt.ylabel('Experience')
plt.title('Experience Prediction')
plt.legend()
plt.show()
```

<h3>Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
iris = load_iris()
```

```python
X = iris.data[:, :]
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
logistic_pred = logistic.predict(X_test)
```

```python
print(accuracy_score(y_test, logistic_pred))
print(confusion_matrix(y_test, logistic_pred))
print(classification_report(y_test, logistic_pred))
```

<h2>Clustering</h2>

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples = 215, n_features = 4, centers = 5,)
X_scaled = scaler.fit_transform(X)
```

```python
km = KMeans(n_clusters = 3)
km.fit(X_scaled)
km_labels = km.predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=km_labels, cmap='viridis', marker='o')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

<h3>DBSCAN</h3>

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(min_samples = 5, eps = 0.3)
dbscan_label = dbscan.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_label, cmap='viridis', marker='o')
plt.title('DBSCAN')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

<h3>Hierarchical</h3>

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X_scaled)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=agglo_labels, cmap='viridis', marker='o')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

<h3>Select K Best</h3>

<h4>Regression task</h4>

```python
from sklearn.feature_selection import SelectKBest, chi2, f_regression
from sklearn.datasets import fetch_california_housing
```

```python
select_reg = SelectKBest(score_func = f_regression, k = 4)  # "f_regression" for Regression task
```

```python
california = fetch_california_housing()
X = pd.DataFrame(california.data, columns=california.feature_names)
y = california.target
```

```python
X.head()
```

```python
X_new = select_reg.fit_transform(X, y)
```

```python
selected_features = X.columns[select_reg.get_support()]
print("Selected Features for Regression:", selected_features.tolist())
```
