```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mp
```

<h3>Sactter Plot</h3>

```python
fmri = seaborn.load_dataset("fmri")
 
seaborn.scatterplot(x="timepoint",y="signal",data=fmri)

plt.show()
```

```python
data = pd.read_csv("football.csv")
```

```python
data.head()
```

```python
sns.scatterplot(x = 'Shots', y = 'Matches_Played', data = data, hue = 'Year',size = 'League', legend = False)
plt.show()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(iris, x ='sepal_length', y = 'sepal_width', hue = 'species', style = 'species', size = 'species', palette='pastel')
plt.show()
```

```python
sns.scatterplot(iris, x ='sepal_length', y = 'sepal_width', hue = 'species', size = 'species', palette='Greys', alpha = .6)
plt.show()
```

```python
sns.scatterplot(data, x = 'Country', y = 'Goals', hue = 'Year', style = 'Year')
plt.show()
```

<h3>Regression Line</h3>

```python
sns.regplot(iris, x = 'sepal_length', y = 'petal_length', color = 'black')
```

```python
sns.lmplot(iris, x = 'sepal_length', y = 'petal_length')
```

<h3>Joint Plot - Marginal Histogram with Scatter plot</h3>

```python
sns.jointplot(iris, x ='sepal_length', y = 'sepal_width',kind = 'reg')
plt.show()
```

```python
sns.jointplot(iris, x ='sepal_length', y = 'sepal_width',hue = 'species')
plt.show()
```

<h3>Line Plot</h3>

```python
sns.lineplot(data, x = 'Matches_Played', y = 'Goals') 
plt.show()
```
